```python
missing_values = df.isnull().sum()
missing_values
```

```python
sorted_missing_values = missing_values.sort_values(ascending=False)
print(sorted_missing_values)
```

```python
# Define the threshold percentage
threshold = 50  # percentage

# Calculate the percentage of missing values for each column
missing_percentage = df.isnull().mean() * 100

# Drop columns where the percentage of missing values exceeds the threshold
columns_to_drop = missing_percentage[missing_percentage > threshold].index
df_cleaned = df.drop(columns=columns_to_drop)

print("Columns dropped:", columns_to_drop)
print(df_cleaned)
```

```python
df_cleaned.isnull().sum()
```

```python
df_cleaned['iso_code'] = df_cleaned['iso_code'].fillna('Not Available')
df_cleaned.isnull().sum()
```

```python
# Identify duplicate rows
duplicates = df_cleaned[df_cleaned.duplicated()]

print("Duplicate rows:")
print(duplicates)
```

```python
# Separate numerical and non-numerical columns
num_columns = df_cleaned.select_dtypes(include='number').columns
non_num_columns = df_cleaned.select_dtypes(exclude='number').columns

# Calculate means for numerical columns
mean_values = df_cleaned[num_columns].mean()

# Fill missing values in numerical columns with their means
df_filled = df_cleaned[num_columns].fillna(mean_values)

# For non-numerical columns, you might use the mode or a specific value
mode_values = df_cleaned[non_num_columns].mode().iloc[0]
df_filled_non_num = df_cleaned[non_num_columns].fillna("Not available")

# Combine the filled numerical and non-numerical columns
df_filled = pd.concat([df_filled, df_filled_non_num], axis=1)

print("DataFrame with filled missing values (numerical and non-numerical columns):")
print(df_filled)
```

```python
# Drop duplicates and save the cleaned DataFrame to a CSV file
df_no_duplicates = df_filled.drop_duplicates()
df_no_duplicates
```

```python
# Save to CSV
df_no_duplicates.to_csv('cleaned_data.csv', index=False)
```
